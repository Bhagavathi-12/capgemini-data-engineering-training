```sql
create database week;
```

```sql
use week;
```

```sql
CREATE TABLE Employees (
    emp_id INT PRIMARY KEY,
    name VARCHAR(50),
    department_id INT,
    salary INT,
    manager_id INT
);
```

```sql
CREATE TABLE Departments (
    dept_id INT PRIMARY KEY,
    dept_name VARCHAR(50)
);
```

```sql
CREATE TABLE Orders (
    order_id INT PRIMARY KEY,
    emp_id INT,
    order_amount INT,
    order_date DATE
);
```

```sql
SELECT *
FROM Employees
WHERE salary > (
    SELECT AVG(salary) FROM Employees
);
```

```sql
SELECT *
FROM Employees
WHERE department_id = (
    SELECT department_id
    FROM Employees
    WHERE name = 'Alice'
);
```

```sql
SELECT *
FROM Employees
WHERE salary = (
    SELECT MIN(salary) FROM Employees
);
```

```sql
SELECT *
FROM Employees e
WHERE EXISTS (
    SELECT 1
    FROM Orders o
    WHERE o.emp_id = e.emp_id
);
```

```sql
SELECT *
FROM Employees
WHERE salary > (
    SELECT MIN(salary)
    FROM Employees
    WHERE department_id = 102
);
```

```sql
WITH high_salary AS (
    SELECT *
    FROM Employees
    WHERE salary > 50000
)

SELECT *
FROM high_salary;
```

```sql
WITH dept_avg AS (
    SELECT 
        department_id,
        AVG(salary) AS avg_salary
    FROM Employees
    GROUP BY department_id
)

SELECT *
FROM dept_avg;
```

```sql
WITH emp_dept AS (
    SELECT 
        e.emp_id,
        e.name,
        e.salary,
        d.dept_name
    FROM Employees e
    JOIN Departments d
    ON e.department_id = d.dept_id
)

SELECT *
FROM emp_dept;
```

```sql
WITH emp_orders AS (
    SELECT 
        emp_id,
        SUM(order_amount) AS total_orders
    FROM Orders
    GROUP BY emp_id
)

SELECT *
FROM emp_orders;
```

```sql
WITH dept_avg AS (
    SELECT 
        department_id,
        AVG(salary) AS avg_salary
    FROM Employees
    GROUP BY department_id
)

SELECT e.*
FROM Employees e
JOIN dept_avg d
ON e.department_id = d.department_id
WHERE e.salary > d.avg_salary;
```
