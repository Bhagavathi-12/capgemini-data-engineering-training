```sql
%sql
Select 
row_number() over( order by salary desc) as rn,
* 
from emp
```

```sql
%sql
Select 
row_number() over( partition by department order by salary desc) as rn,
* 
from emp
```

```sql
%sql
Select 
row_number() over( order by join_date ) as rn,
* 
from emp
```

```sql
%sql
Select 
row_number() over( order by join_date DESC ) as rn,
* 
from emp
```

```sql
%sql
Select 
row_number() over( order by salary  ) as rn,
* 
from emp
```

```sql
%sql
Select 
row_number() over(Partition by department order by emp_name  ) as rn,
* 
from emp
```

```sql
%sql
Select 
rank() over( order by salary desc  ) as rn,
* 
from emp
```

```sql
%sql
Select 
rank() over(Partition by department order by salary ) as rn,
* 
from emp
```

```sql
%sql
Select 
rank() over(Partition by department order by join_date DESC ) as rn,
* 
from emp
```

```sql
%sql
Select 
rank() over(order by emp_name) as rn,
* 
from emp
```
