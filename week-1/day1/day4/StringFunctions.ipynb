### LENGTH / CHAR_LENGTH

```sql
SELECT full_name, LENGTH(full_name) FROM employees_data3;
SELECT full_name, CHAR_LENGTH(full_name) FROM employees_data3;
```

### UPPER / LOWER

```sql
SELECT UPPER(department), LOWER(city) FROM employees_data3;
```

### TRIM / LTRIM / RTRIM

```sql
SELECT
TRIM(full_name),
LTRIM(full_name),
RTRIM(full_name)
FROM employees_data3;
```

### CONCAT

```sql
SELECT CONCAT(full_name, ' - ', department) AS emp_details from employees_data3
```

### CONCAT_WS (With Separator)

```sql
SELECT CONCAT_WS(' | ', emp_id, full_name, city) from employees_data3;
```

### SUBSTRING / SUBSTR / MID

```sql
SELECT SUBSTRING(email, 1, 7) FROM employees_data3;
```

```sql
SELECT SUBSTR(email, 1, 5) FROM employees_data3;
```

### LEFT / RIGHT

```sql
SELECT LEFT(full_name, 4), RIGHT(city, 3)
FROM employees_data3;
```

### INSTR - Position of substring

```sql
SELECT email, INSTR(email, '@') FROM employees_data3;
```

### LOCATE - Same as INSTR but with more flexibility


```sql
SELECT email, LOCATE('.', email) FROM employees_data3;
```

### REPLACE - Replace substring


```sql
SELECT REPLACE(department, 'Data', 'Big Data') from employees_data3;
```

### REVERSE - Reverse a string


```sql
SELECT full_name, REVERSE(full_name) from employees_data3
```

### LPAD / RPAD - Pad characters on left or right


```sql
SELECT LPAD(emp_id, 5, '0') FROM employees_data3;
SELECT RPAD(city, 15, '*') FROM employees_data3;
```

### TRIM + REPLACE (Combined Usage)

```sql
SELECT TRIM(REPLACE(city, ' ', '')) FROM employees_data3;
```

### IFNULL - Handle NULL strings


```sql
SELECT full_name, IFNULL(remarks, 'No remarks')
FROM employees_data3;
```

### COALESCE - Return first non-NULL value

```sql
SELECT full_name, COALESCE(remarks, 'N/A') from employees_data3;
```

### FIND_IN_SET - Search value in a comma-separated list (CSV)


```sql
SELECT FIND_IN_SET('Analytics', 'Data,Analytics,AI');
```
