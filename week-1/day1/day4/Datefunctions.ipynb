```sql
CREATE TABLE orders3 (
order_id INT PRIMARY KEY,
customer_name VARCHAR(50),
order_date DATE,
order_timestamp TIMESTAMP,
delivery_date DATE,
order_amount DECIMAL(10,2)
);

INSERT INTO orders3 VALUES
(1, 'Karthik', '2024-01-15', '2024-01-15 10:30:45', '2024-01-20', 2500.00),
(2, 'Veena', '2024-02-18', '2024-02-18 18:45:20', '2024-02-22', 3200.50),
(3, 'Ravi', '2024-03-02', '2024-03-02 09:15:10', '2024-03-08', 4100.75),
(4, 'Anil', '2024-03-09', '2024-03-09 14:05:55', '2024-03-15', 1800.00),
(5, 'Suresh', '2024-01-07', '2024-01-07 23:55:00', '2024-01-12', 2900.00);
```

EXTRACTING DATE PARTS

```sql
SELECT YEAR(order_date), MONTH(order_date), DAY(order_date)
FROM orders3;
```

```sql
SELECT
EXTRACT(YEAR FROM order_date),
EXTRACT(MONTH FROM order_date),
EXTRACT(DAY FROM order_date)
FROM orders3;
```

MONTH & DAY NAMES

```sql
SELECT MONTHNAME(order_date), DAYNAME(order_date)
FROM orders3;
```

```sql
SELECT
WEEKDAY(order_date),
DAYOFWEEK(order_date)
FROM orders3;
```

IDENTIFY WEEKENDS (SATURDAY & SUNDAY)

```sql
SELECT order_id, order_date
FROM orders3
WHERE DAYNAME(order_date) IN ('Sat', 'Sun');
```

```sql
SELECT order_id, order_date
FROM orders3
WHERE DAYOFWEEK(order_date) IN (1, 7);
```

IDENTIFY WEEKDAYS

```sql
SELECT order_id, order_date
FROM orders3
WHERE DAYOFWEEK(order_date) BETWEEN 2 AND 6;
```

DATE ARITHMETIC

```sql
SELECT order_date, DATE_ADD(order_date, 3)
FROM orders3;
```

```sql
SELECT order_date, DATE_SUB(order_date, 3)
FROM orders3;
```

```sql
SELECT add_months(order_date, 1) FROM orders3;
SELECT add_months(order_date, -2) FROM orders3;
```

DATEDIFF (Difference in Days)

```sql
SELECT
order_id,
DATEDIFF(delivery_date, order_date) AS delivery_days
FROM orders3;
```

```sql
SELECT
TIMESTAMPDIFF(DAY, order_date, delivery_date) AS days_diff,
TIMESTAMPDIFF(MONTH, order_date, delivery_date) AS months_diff
FROM orders3;
```

```sql
SELECT LAST_DAY(order_date) FROM orders3
```

FIRST DAY OF MONTH

```sql
SELECT DATE_SUB(order_date, DAY(order_date)-1) from orders3
```

Orders in Last 7 Days

```sql
SELECT *
FROM orders3
WHERE order_date >= CURDATE() - INTERVAL 7 DAY;
```

Orders Placed Today

```sql
SELECT *
FROM orders3
WHERE DATE(order_timestamp) = CURDATE();
```
